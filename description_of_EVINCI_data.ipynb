```python
import os
import pickle
import matplotlib.pyplot as plt

#location of file
fp = os.path.join('data','anon','001.pickle')

#load dictionary
with open(fp,'rb') as f:
    im = pickle.load(f)
```

```python
#what have we unpickled? should be a dictionary
print(type(im))
```

```python
#what kesy does this dictionary have?
print(im.keys())
```

```python
#what type does each of the values have?
print({k:type(v) for k,v in zip(im.keys(),im.values())})
```

```python
#what does ths pixel spacing look like? should be 2 numeric elements
print(im['pxSpacing'])
```
